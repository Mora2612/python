```python
import pandas as pd

url = 'https://raw.githubusercontent.com/softhints/Pandas-Exercises-Projects/main/data/food_recipes.csv'
df = pd.read_csv(url)

print("=== Informacje o DataFrame ===")
df.info()

print("\n=== Statystyki opisowe ===")
print(df.describe(include='all'))

print("\n=== Liczba brakujących wartości w każdej kolumnie ===")
print(df.isnull().sum())

df_clean = df.drop_duplicates()

print("\nLiczba wierszy przed czyszczeniem:", len(df))
print("Liczba wierszy po czyszczeniu:", len(df_clean))

print("\n=== Podgląd oczyszczonych danych ===")
print(df_clean.head())
```
