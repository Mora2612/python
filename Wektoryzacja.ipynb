```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "amount": [500, 1500, 2000, 300],
    "country": ["PL", "PL", "DE", "PL"],
    "discount": [0.1, 0.2, 0.0, 0.15]
})
```

```python
condition = (df["country"] == "PL") & (df["amount"] > 1000)
```

```python
df["final_price"] = np.where(
    condition,
    df["amount"] * (1 - df["discount"]), # Jeśli prawda
    df["amount"]                         # Jeśli fałsz
)
```
