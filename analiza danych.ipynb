```python
import pandas as pd

data = { 
    "Imię": ["Ala", "Bartek", "Celina"], 
    "Wiek": [22, 25, 20], 
    "Ocena": [4.5, 3.8, 5.0] 
} 

df = pd.DataFrame(data)

display(df)

print(df.describe())

df["Rok studiów"] = [1, 2, 1]

print("\nPo dodaniu kolumny 'Rok studiów':")
display(df)

srednia = df["Ocena"].mean()
print(f"\nŚrednia z kolumny 'Ocena': {srednia:.2f}")

df_ocena_wyzsza = df[df["Ocena"] > 4]

print("\nWiersze, w których ocena > 4:")
display(df_ocena_wyzsza)
```
