```python
import numpy as np
import pandas as pd
```

```python
def process_sales_optimized(path):
    przetworzone_chunki = []

    for chunk in pd.read_csv(path, chunksize=50000):
        chunk["flag"] = np.where((chunk["amount"] > 1000) & (chunk["country"] == "PL"), 1, 0)
        chunk["country"] = chunk["country"].astype("category")
        if "product" in chunk.columns:
            chunk["product"] = chunk["product"].astype("category")
        chunk["amount"] = pd.to_numeric(chunk["amount"], downcast="integer")

        przetworzone_chunki.append(chunk)
    return pd.concat(przetworzone_chunki, ignore_index=True)
```

```python
optymdf = process_sales_optimized("duzy_plik.csv")


print(optymdf.info())
```
